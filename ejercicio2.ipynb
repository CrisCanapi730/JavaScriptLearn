```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
t = np.linspace(0,10,100)
t
```

```python
L = 1 - 1/2*np.exp(-2*t)
L
```

```python
pl.plot(t,L)
pl.grid()
```

```python
L_1 = 1-0.5*np.exp(-2*t)
L_2 = 1-np.exp(-2*t)
pl.plot(t, L_1, label = 'sabe la mitad')
pl.plot(t, L_2, label = 'no sabe nada') # t es time o tiempo
pl.grid()
pl.legend()
```
